```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
p    = 0.9
mean = [0,0]
cov  = [[1,p],[p,1]]
n    = 100

x1, x2 = np.random.multivariate_normal(mean, cov, n).T
```

```python
point = (-1.7, -1.7)
```

```python
m = 15

x2_cond = np.random.normal(loc=point[0], scale=np.sqrt(1-p**2), size=m)
x2_marg = np.random.choice(x2, size=m)
```

```python
ax1 = plt.subplot(121)
plt.scatter(x1, x2, color='black', alpha=0.1)
plt.scatter(np.repeat(point[0], m), x2_marg, color='blue')
plt.scatter(point[0], point[1], color='red')
plt.ylabel('x2')
plt.xlabel('x1')
plt.subplot(122, sharey=ax1)
plt.scatter(x1, x2, color='black', alpha=0.1)
plt.scatter(np.repeat(point[0], m), x2_cond, color='green')
plt.scatter(point[0], point[1], color='red')
plt.xlabel('x1')
plt.show()
```
