```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mu, sigma, samples = 2, 0.5, 1000000 # mean, standard deviation and number of samples
data = np.random.normal(mu, sigma, samples)
```

```python
x_min, x_max = np.floor(data.min()), np.ceil(data.max())

bar_number = 500.0
bar_width  = (x_max - x_min) / bar_number
```

```python
x_values, cdf_values, pdf_values = [], [], []
cdf_last = 0

for x in np.arange(x_min, x_max, bar_width):
    cdf      = len( [d for d in data if d < x ] ) / samples
    pdf      = (cdf - cdf_last) / bar_width
    cdf_last = cdf
    x_values.append(x)
    pdf_values.append(pdf)
    cdf_values.append(cdf)
   
```

```python
plt.title('Probability Density Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.scatter(x_values, cdf_values, label='cdf', marker='.')
plt.scatter(x_values, pdf_values, label='pdf', marker='.')
plt.legend()

# Show the plot
plt.show()
```

```python
# Expectation (mean) and Variance
E = sum(data)/len(data)
V = sum((data - E)**2) / len(data)
std = np.sqrt(V)
```
