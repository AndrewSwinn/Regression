```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x  = np.random.rand(100) * 10
y  = x * 2 +  np.random.rand(100) * 10
X  = np.stack((x, y), axis=1)
Xc = X - X.mean(axis=0)
```

```python
plt.scatter(Xc[:,0], Xc[:,1])
plt.show()
```
